# To Display Value Count For Two Categorical Variables

# Importing The Required Library

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

# Load The Data

```python
df = pd.read_csv("E:\Project_Work\Test_data.csv")
```

```python
df.describe()
```

```python
df.head()
```

# Count Plot On Two Categorical Variables

```python
sns.countplot(x ='gender', hue = "department", data = df)
```

# Count Plot On Two Categorical Variables

```python
sns.countplot(x ='gender', hue = "education", data = df)
```

# To Display A Boxplot Using Only One Variable

```python
sns.set(style='whitegrid')
sns.boxplot(x='age', data=df)
```

# To Display A Boxplot Using Only One Variable

```python
sns.set(style='whitegrid')
sns.boxplot(x='avg_training_score', data=df)
```

```python
# Histogram for numerical variable
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.hist(df['length_of_service'], bins=20, color='skyblue')
plt.xlabel('Length of Service')
plt.ylabel('Frequency')
plt.title('Distribution of Lenth of Service')
plt.show()
```
